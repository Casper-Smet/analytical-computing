# Planes in $R^{3}$

```python
%matplotlib inline
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
point  = np.array([1, 2, 3])
normal = np.array([1, 1, 2])
point2 = np.array([10, 50, 50])
d = -point.dot(normal)
xx, yy = np.meshgrid(range(10), range(10))
z = (-normal[0] * xx - normal[1] * yy - d) * 1. /normal[2]
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(xx, yy, z, alpha=0.2)
plt.show()
```


Equation of a plane in $R^{3}$:<BR>
$ax+by+cz=D$ <br>
Another way to specify a plane, is by giving a point on the plane. <br>
$\vec{n} \Rightarrow$ "Normal" vector, perpendicular to everything on the plane. <br>
Only when $\vec{n}\cdot\vec{a}=0$, is $\vec{a}$ perpendicular to $\vec{n}$ (see angles-and-vectors.ipynb).<br>
$\vec{n} + (x_{0},y_{0},z_{0})$ <br>
$\vec{x}_{0} = \begin{bmatrix} x_{0}\\y_{0}\\z_{0}\end{bmatrix},\vec{x}=\begin{bmatrix}x\\y\\z\end{bmatrix}$ <br>
$(\vec{x} - \vec{x}_{0}) \Leftarrow $ perpendicular to $\vec{n}$ <br>
$\vec{n}=\begin{bmatrix}n_{1}\\n_{2}\\n_{3}\end{bmatrix},\space \vec{n}\cdot(\vec{x} - \vec{x}_{0}) = 0$<br>
$\begin{bmatrix}n_{1}\\n_{2}\\n_{3}\end{bmatrix}\cdot \begin{bmatrix} x-x_{0}\\y-y_{0}\\z-z_{0}\end{bmatrix}=0$ <br>

$n_{1}(x-x_{0})+ n_{2}(y-y_{0}) + n_{3}(z-z_{0})=0 \qquad \Rightarrow \qquad ax+by+cz=D$

## Proving this formula (twice):
For the following vectors:
$\vec{n} = \begin{bmatrix}1\\3\\-2\end{bmatrix},\space\vec{x}_{0}=\begin{bmatrix} 1\\2\\3 \end{bmatrix},\space
\vec{x}=\begin{bmatrix}x\\y\\z\end{bmatrix}$ <br>
### Proof 1
$\begin{bmatrix}1\\3\\-1\end{bmatrix}\cdot\begin{bmatrix}x-1\\y-2\\z-3\end{bmatrix}=0$ <br>
$(x-1) + 3(y-2)-2(z-3) = 0$<br>
>$x-1+3y-6-2z + 6= 0$ <br>
$x-1+3y-2z=0$ 

$x+3y-2z=1$

### Proof 2
$n_{1}(x-x_{0})+ n_{2}(y-y_{0}) + n_{3}(z-z_{0})=0$<br>
$1(x-1)+3(y-2)-2(z-3)=0$ <br>
$x-1 + 3y - 6 -2z + 6=0$ <br>
>$x-1+3y-2z=0$

$x+3y-2z=1$
